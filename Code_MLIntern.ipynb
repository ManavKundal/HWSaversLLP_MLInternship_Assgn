```python

def whether_capital(df):
    
    a = 0
    b = []
    for skill in df:
        a = 0
        for word in skill.split():
            if word[0].islower():
                a+=1
        if (a > 0):
            b.append(0)
        else:
            b.append(1)
        
    return b
```

```python
req = whether_capital(df['RAW DATA'])
print(req)
```

```python
req_df = df
req_df['Tech_Skills'] = req
```
